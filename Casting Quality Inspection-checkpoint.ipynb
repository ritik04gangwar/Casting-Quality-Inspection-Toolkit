```python
# Converting the image list and label list to numpy arrays
X_test = np.array(data)
Y_test = np.array(labels)
print(X_test.shape, Y_test.shape)
```

```python
# Scaling the data
X_test_scaled = X_test/255
```

```python
import tensorflow as tf
from tensorflow import keras
```

```python
import numpy as np
from tensorflow import keras
from tensorflow.keras import regularizers

# Define number of classes
number_of_classes = 5

# Build the model
model = keras.Sequential([
    keras.layers.Conv2D(32, kernel_size=(3,3), activation='relu', input_shape=(224,224,3),
                        kernel_regularizer=regularizers.l1(0.01)),
    keras.layers.MaxPooling2D(pool_size=(2,2)),
    keras.layers.Conv2D(64, kernel_size=(3,3), activation='relu',
                        kernel_regularizer=regularizers.l1(0.01)),
    keras.layers.MaxPooling2D(pool_size=(2,2)),
    keras.layers.Flatten(),
    keras.layers.Dense(128, activation='relu',
                       kernel_regularizer=regularizers.l1(0.01)),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(number_of_classes, activation='softmax',
                       kernel_regularizer=regularizers.l1(0.01))
])

# Print the model summary
model.summary()
```

```python
# Compile the model
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
# Train the model
history = model.fit(X_train_scaled, Y_train, validation_split=0.1, epochs=10)
```

```python
 loss, accuracy = model.evaluate(X_test_scaled, Y_test)
print('Test Accuracy: ', accuracy)
```

```python
h = history

# plot the loss value
plt.plot(h.history['loss'], label='train loss')
plt.plot(h.history['val_loss'], label='val loss')
plt.legend()
plt.show()

# plot the accuracy value
plt.plot(h.history['accuracy'], label='train acc')
plt.plot(h.history['val_accuracy'], label='val acc')
plt.legend()
plt.show()
```

```python
# Predictive System
def load_and_preprocess_image(input_image_path,target_size=(224,224)):
    input_image = cv2.imread(input_image_path)
    input_image_resized = cv2.resize(input_image, (224, 224))
    input_image_scaled = input_image_resized / 255
    input_image_reshaped = np.reshape(input_image_scaled, [1, 224, 224])

def predict_image_class(model,image_path,class_indices):
    preprocessed_img=load_and_preprocess_image(image_path);
    input_prediction = model.predict(input_image_reshaped)
    input_pred_label = np.argmax(input_prediction,axis=1)[0]
    pred_class_name=class_indices[input_pred_label]
    return pred_class_name
```
